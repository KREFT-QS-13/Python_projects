# Bloch Sphere and Matrixes

```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector as pbm
from qiskit.visualization import plot_histogram as ph
```

```python
circ = QuantumCircuit(1,1)
circ.x(0)
circ.h(0)

%matplotlib inline
circ.draw(output='mpl')
```

```python
simulator_u = Aer.get_backend('unitary_simulator')
result = execute(circ, backend=simulator_u).result()
unitary = result.get_unitary()

simulator_v = Aer.get_backend('statevector_simulator')
result = execute(circ, backend=simulator_v).result()
statevector = result.get_statevector()

print("Matrix for X gate:\n {}".format(unitary))
print("\nVector for result:\n {}".format(statevector))
```

```python
pbm(statevector)
```

```python
circ2 = QuantumCircuit(1,1)
circ2.x(0)
circ2.h(0)
circ2.measure([0], [0])

simulator = Aer.get_backend('qasm_simulator')
result = execute(circ2, backend=simulator, shots=1024).result()
counts = result.get_counts()

ph(counts)
```
