# Quantum Teleportation Algorithm

```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram as plt_hist
```

```python
circ = QuantumCircuit(3,3)
# we want to teleport state of q0 to q2
circ.x(0)
circ.barrier()

# Teleportaion protocol
circ.h(1)
circ.cx(1,2)
circ.cx(0,1)
circ.h(0)

circ.barrier()
circ.measure([0,1],[0,1])

circ.barrier()
circ.cx(1,2)
circ.cz(0,2)
```

```python
%matplotlib inline
circ.draw(output='mpl')
```

```python
# Verify the output
circ.measure(2,2)
sim = Aer.get_backend('qasm_simulator')
result = execute(circ, backend = sim, shots = 1024).result()
counts = result.get_counts()
plt_hist(counts)
```
