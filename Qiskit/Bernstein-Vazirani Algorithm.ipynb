# Bernstein-Vazirani Algorithm

```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram as plth
%matplotlib inline
```

```python
secretnumber = '1100'
```

```python
circ = QuantumCircuit(len(secretnumber)+1, len(secretnumber))

circ.h(range(len(secretnumber)))
circ.x(len(secretnumber))
circ.h(len(secretnumber))
circ.barrier()

for i, yesno in enumerate(reversed(secretnumber)):
    if yesno == '1':
        circ.cx(i, len(secretnumber))

circ.barrier()

circ.h(range(len(secretnumber)))
circ.barrier()
circ.measure(range(len(secretnumber)), range(len(secretnumber)));
```

```python
circ.draw(output='mpl')
```

```python
sim = Aer.get_backend('qasm_simulator')
result = execute(circ, backend=sim, shots=1).result()
count = result.get_counts()
```

```python
# Checking the answears
for x in count.keys():
    print( int(x), int(secretnumber) )
```
